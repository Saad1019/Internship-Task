## **Dataset Understanding and Description**

The Iris Dataset contains 150 samples of iris flowers with 4 features:






● Sepal length (cm)



● Sepal width (cm)



● Petal length (cm)



● Petal width (cm) It also includes the species (setosa, versicolor, virginica). The dataset is clean with no missing values.

## **Data Cleaning and Preparation**

No cleaning is needed as the dataset has no missing values or inconsistencies.

## **Exploratory Data Analysis (EDA)**

We will create visualizations to understand relationships, distributions, and potential outliers.

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = sns.load_dataset('iris')

# Display dataset structure
print("Shape:", iris.shape)
print("Columns:", iris.columns)
print(iris.head())

```

```python
# Scatter plot: Petal length vs. Petal width by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

```

```python
# Histogram: Distribution of each feature
iris.hist(figsize=(10, 8))
plt.suptitle('Distribution of Iris Features')
plt.show()

```

```python
# Box plot: Spread and outliers for each feature
plt.figure(figsize=(10, 6))
iris.boxplot()
plt.title('Box Plot of Iris Features')
plt.show()
```

## **Conclusion**

● The scatter plot shows that petal length and width are strongly correlated, with clear separation between species (e.g., setosa has smaller petals).



● Histograms reveal that petal length and width have multimodal distributions, indicating distinct groups (species).
